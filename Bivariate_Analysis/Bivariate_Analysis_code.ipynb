```python
import pandas as pd
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
heart = pd.read_csv('heart.csv')
```

```python
heart
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

## 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(data=iris,
                x='sepal_length',
                y='sepal_width',
                hue='species',
                style='species',
                size='petal_width')
```

## 2. Barplot (Numerical - Categorical)

```python
sns.barplot(data = titanic,
            x='class',
            y='fare',
            hue='sex')
```

## 3. Box plot (Numerical - Categorical)

```python
sns.boxenplot(data = titanic,
              x='sex',
              y='age',
              hue='alive')
```

## 4. Distplot (Numerical - Categorical)

```python
sns.distplot(heart[heart['target']==0]['age'], hist=False)
sns.distplot(heart[heart['target']==1]['age'], hist=False)
```

## 5. HeatMap (Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(heart['cp'], heart['target']), annot=True)
```

## 6. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['embark_town'],
                           titanic['survived']))
```

## 7. Pairplot

```python
sns.pairplot(heart[['age','chol','thalach','trestbps','target']],
             hue='target')
```

## 8. Lineplot (Numerical - Numerical)

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
sns.lineplot(data=flights,
             x='year',
             y='passengers',
             hue='month')
```

## 9. ClusterMap

```python
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
