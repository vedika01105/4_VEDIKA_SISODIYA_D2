Import Libraries

```python
from sklearn.datasets import make_regression
import numpy as np
```

Generate Dataset

```python
X, y = make_regression(n_samples=4,
                       n_features=1,
                       n_informative=1,
                       n_targets=1,
                       noise=80,
                       random_state=13)
```

Plot the Dataset

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], y)
```

Apply OLS (Ordinay Least Squares)

```python
from sklearn.linear_model import LinearRegression
```

Train The Model

```python
reg = LinearRegression()
reg.fit(X,y)
```

Check Slope

```python
reg.coef_
```

Check Intercept

```python
reg.intercept_
```

Plot the Regression line

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

Start Gradient Descent manually

```python
y_pred = ((78.35 * X) + 100).reshape(4)
```

Plot this wrong line

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

Calculate Gradient (Loss slope)

```python
m = 78.35
b = 100
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

Define Learning Rate

```python
lr = 0.1
```

Calculate step size

```python
step_size = loss_slope * lr
step_size
```

Update Intercept

```python
b = b - step_size
b
```

New Prediction

```python
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```
